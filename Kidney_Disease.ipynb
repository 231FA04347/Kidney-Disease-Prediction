```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd

# Data Loading
dt = pd.read_csv('/content/kidney_disease.csv')

# Check if the DataFrame was loaded successfully
if dt is None:
    print("Error: Failed to load the CSV file. 'dt' is None.")
    # You might want to add more specific error handling here,
    # like exiting the script or trying a different file path.
else:
    dt.head()
    dt = dt.dropna()
    dt.info()
```

```python
dt['rbc'].unique()
```

```python
for i in dt.columns:
  if dt[i].dtype=='object':
    dt[i]=dt[i].astype('category').cat.codes
```

```python
dt.info()
```

```python
dt
```

```python
dt.corr()
```

```python
dt = dt.drop(['rbc','rc'],axis=1)
```

```python
x=dt.iloc[:,0:19]
y=dt['classification']
```

```python
from sklearn.model_selection import train_test_split   #Data Split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
obj=Sequential()
```

```python
print(x.shape,y.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
obj.add(Dense(units=19,activation='relu'))  #Model Architecture
```

```python
obj.add(Dense(units=5,activation='relu'))
```

```python
obj.add(Dense(units=1,activation='sigmoid'))
```

```python
obj.build(input_shape=(None,19))
```

```python
obj.summary()
```

```python
obj.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
obj.fit(x_train,y_train,epochs=100,batch_size=8,steps_per_epoch=x_train.shape[0]//8)
```

```python
y_pred = obj.predict(x_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
obj.save('Kidney.h5')
```
